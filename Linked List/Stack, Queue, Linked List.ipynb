```python
#POP into a Stack
class Stack:
   def __init__(self):
      self.stack = []

# Use list append method to add element
   def add(self, dataval):
      if dataval not in self.stack:
         self.stack.append(dataval)
         return True
      else:
         return False
        
# Use list pop method to remove element
   def remove(self):
      if len(self.stack) <= 0:
         return ("No element in the Stack")
      else:
         return self.stack.pop()

AStack = Stack()
AStack.add("Mon")
AStack.add("Tue")
AStack.add("Wed")
AStack.add("Thu")
print(AStack.remove())
print(AStack.remove())
```

### Linked List
A linked list is a sequence of data elements, which are connected together via links. Each data element contains a connection to another data element in form of a pointer. <br> 
1. Singly Linked List
2. Doubly Linked List
3. Circular Linked List <br>

Advantages: <br>
* Dynamic data structure
* Insertion and Deletion is easy
* You can implement stack,queue,graph

Disadvantades: <br>
* Needs extra memory
* Random access not posible

Application: <br>
* Web browser
* Music player: songs are linked
* Image viewer: images are linked

## Circular Linked List

1. Circular Singly Linked List
2. Circular Doubly Linked List

* Insertion (begin, end, inbetween(After, Before))
* Delete (begin, end, inbetween)
* Traversal

A circular linked list is similar to a linked list in that the nodes are connected by links, but the last node is also linked to the first node instead of just linking to nothing. So, after we've accessed the last node, we can access the first node through the last node.
